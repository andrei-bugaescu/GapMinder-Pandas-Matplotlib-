# Atelier Visualisation: Reproduire le "très" célèbre graph du GapMinder avec Pandas and Matplotlib

GapMinder, selon sa présentation officielle, est une organisation non gouvernementale à but non lucratif, dont le but est la promotion du développement durable et des [objectifs du millénaire pour le développement](https://www.who.int/topics/millennium_development_goals/about/fr/#:~:text=Les%20huit%20objectifs%20du%20Mill%C3%A9naire,l'%C3%A9ducation%20primaire%20pour%20tous%3B&text=mettre%20en%20place%20un%20partenariat%20mondial%20pour%20le%20d%C3%A9veloppement.) de l'ONU. La démarche est originale : l’ONG veut y contribuer à sa manière, par la popularisation d’un outil de visualisation statistique. Son créateur, Hans Rosling, professeur et chercheur suédois en médecine et santé publique, a longtemps travaillé sur les problèmes sanitaires dans les pays en voie de développement.

Le principe fondateur de cet outil est l’accès graphique simplifié et interactif à une grande quantité de données statistiques, dans leurs dimensions spatiale et temporelle, et leur croisement à l’aide de graphiques de type «nuages de points».

La présentation par défaut est celle du croisement de l’espérance de vie à la naissance et du PIB par habitants, pour une grande majorité des États du monde. Les pays sont donc placés sur les deux axes selon leurs valeurs, et représentés par un cercle proportionnel à leur population pour l’année donnée. Une quatrième dimension d’information est apportée par la couleur de ces symboles qui, par défaut, signifie l’appartenance du pays à une une région du monde (<i>[source](http://mappemonde-archive.mgm.fr/num18/internet/int08201.html)</i>).

Rendez-vous sur [gapminder.com](https://www.gapminder.org/).
